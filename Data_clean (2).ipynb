```python
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
f = open('barcode_midslow','r')
```

```python
data = []
for i in f:
    if i[0:5] == 'data:':
        v = re.split(' ',i)
        data.append(int(v[4][:-1]))
```

```python
plt.plot(data,'r*')
```

```python
plt.plot(data[0:500],'green')
```
